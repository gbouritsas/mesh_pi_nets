```python
import matplotlib.pyplot as plt
# line 1 points
# non-linear 
x = [1, 2, 3, 4]
y1 = [0.635, 0.528, 0.507, 0.501]
y2 = [0.635, 0.476, 0.391, 0.365]
y3 = [0.635, 0.476,0.393, 0.357]


# linear 
y4 = [0.852, 0.533, 0.524, 0.563]
y5 = [0.852, 0.473, 0.395, 0.369]
y6 = [0.852, 0.473,0.384, 0.356]



plt.figure(figsize=(15,10))
# [14.293, 14.038, 13.974]
# plotting the line 1 points 
plt.xticks(x, [1,2,3,4], fontsize = 18)
plt.yticks( fontsize = 18)
plt.plot(x, y1,  linestyle='--', marker='o', markersize=8, label = "ProdPoly simple, non-linear ")
plt.plot(x, y2,  linestyle='--', marker='o', markersize=8,label = "ProdPoly full, non-linear")
plt.plot(x, y3, linestyle='--', marker='o', markersize=8,label = "ProdPoly full, non-linear, residual ")
plt.plot(x, y4, linestyle='--', marker='o', markersize=8,label = "ProdPoly simple, linear")
plt.plot(x, y5, linestyle='--', marker='o', markersize=8,label = "ProdPoly full, linear ")
plt.plot(x, y6, linestyle='--', marker='o', markersize=8,label = "ProdPoly full, linear, residual")

plt.plot(1, 0.732 ,'*', markersize=12, label = "GAT")
plt.plot(1, 0.623 ,'d', markersize=12, label = "FeastNet")
plt.plot(1, 0.583 ,'+r', markersize=12, label = "MoNet")
plt.plot(1, 0.635 ,'hg', markersize=12, label = "SpiralGNN")
plt.plot(1, 0.751 ,'pb', markersize=12, label = "COMA")
plt.plot(1, 0.825 ,'Pm', markersize=12, label = "PCA")

plt.xlabel('order', fontsize=18)
plt.ylabel('generalization error', fontsize=18)
plt.title('COMA dataset', fontsize=18)
plt.legend(prop={'size': 16})
plt.grid()
plt.show()

plt.savefig('/homes/gb318/Downloads/fig1.png', bbox_inches='tight')
```
